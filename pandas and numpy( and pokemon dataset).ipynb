# Aggregate Value (grouping by)


```python
df3 = pd.read_csv("pokemon.csv.txt")
df3
```

```python
df3.groupby(['Type 1']).mean()
```

```python
df3.groupby(['Type 1']).mean().sort_values('Defense')
```

```python
df3.groupby(['Type 1']).mean().sort_values('Defense', ascending= False)
```

```python
df3.groupby(['Type 1']).sum()
```
